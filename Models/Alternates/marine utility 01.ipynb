```python
loc = data['Location']
cfs = data['Common Fish Species']
fr = data['Fishing Regulations']
lat = data['Latitude']
longi = data['Longitude']
```

```python
from math import radians, sin, cos, sqrt, atan2

values = pd.DataFrame({
    'Location': loc,
    'Common Fish Species': cfs,
    'Fishing Regulations': fr,
    'Latitude': lat,
    'Longitude': longi
})

def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c
    return distance

def find_nearest_places(reference_place, places, fish_species):
    reference_lat, reference_lon = reference_place['Latitude'], reference_place['Longitude']
    places_with_distances = []
    visited_locations = set()

    fish_species = fish_species.lower()

    for place in places:

        if fish_species in place['Common Fish Species'].lower():
            distance = haversine(reference_lat, reference_lon, place['Latitude'], place['Longitude'])
            location_name = place['Location']


            if location_name not in visited_locations:
                places_with_distances.append((location_name, distance))
                visited_locations.add(location_name)

    places_with_distances.sort(key=lambda x: x[1])


    return places_with_distances[1:7]
```

```python
print('Enter the details :\n')

def get_lat_lon_from_location(user_location):

    location_row = data[data['Location'] == user_location]

    if location_row.empty:
        return None, None
    else:
        latitude = location_row['Latitude'].iloc[0]
        longitude = location_row['Longitude'].iloc[0]
        return latitude, longitude

Loc = input('Please enter your location :')
Lat, Long = get_lat_lon_from_location(Loc)

if Lat != None or Long != None:
  reference_place = {'Location': Loc, 'Latitude': Lat ,'Longitude': Long}

  fish_species = input('Kind of Specie :')

  nearest_places = find_nearest_places(reference_place, values.to_dict(orient='records'), fish_species)
  print()
  print()

  count=0
  for place, distance in nearest_places:
    if count==0:
      print('Near you: ')
    print(f'{place}, Distance: {distance:.2f} km')
    count=count+1

  if count==0:
    print('Invalid Fish Specie')

else:
  print('Invalid Location')
```
